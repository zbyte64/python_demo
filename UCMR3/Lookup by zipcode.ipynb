```python
#display a series
results['Size'].head()
```

```python
zipcode_table = pd.read_csv('UCMR3_ZipCodes.txt', sep='\t')
zipcode_table.head()
```

```python
zipcode_to_PWSID = lambda x: zipcode_table[zipcode_table['ZIPCODE'] == x]['PWSID'].values[0]
```

```python
zipcode_table[zipcode_table['ZIPCODE'] == '93728']#['PWSID'].values[0]
```

```python
our_pwsid = zipcode_to_PWSID('93728')
our_pwsid
```

```python
#using comparison expressions -> generate bitmask series -> then filter dataframe
filter_location = results['PWSID'] == our_pwsid
filter_equality = results['AnalyticalResultsSign'] == '='
our_rows = results[filter_location & filter_equality]
our_rows
```

```python
filter_location
```

```python

ref_levels = pd.read_csv('epa_regulation_levels.csv', index_col='Contaminant')
ref_levels
```

```python
ref_levels.loc['cobalt', 'Reference Concentration']
```

```python
doselect = lambda x: ref_levels.ix[our_rows.loc[x, 'Contaminant'], 'Reference Concentration'] <= our_rows.loc[x, 'AnalyticalResultValue'] 
exceeded_levels = our_rows.select(doselect)
```
